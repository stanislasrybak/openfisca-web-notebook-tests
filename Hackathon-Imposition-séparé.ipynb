```python
import collections
import json
import urllib2

import numpy as np
```

```python
simulation = dict(
    decomposition = [
        dict(code = 'sali', title = 'Salaire imposable'),
        dict(code = 'irpp', title = 'Impôt sur le revenu des personnes physiques'),
        dict(code = 'revdisp', title = 'Revenu disponible'),
        ],
    scenarios = [
        dict(
             axes = [
                    dict(
                        index = 0,
                        count = 3,
                        max = 100000,
                        min = 0,
                        name = 'sali',
                        ),
                    dict(
                        index = 1,
                        count = 3,
                        max = 100000,
                        min = 0,
                        name = 'sali',
                        ),
                ],
            test_case = dict(
                familles = [
                    dict(
                        parents = ['ind0', 'ind1'],
                        ),
                    ],
                foyers_fiscaux = [
                    dict(
                        declarants = ['ind0'],
                        ),
                    dict(
                        declarants = ['ind1'],
                        ),
                    ],
                individus = [
                    dict(
                        activite = u'Actif occupé',
                        birth = 1970,
                        cadre = True,
                        id = 'ind0',
                        statmarit = u'Célibataire',
                        ),
                    dict(
                        activite = u'Actif occupé',
                        birth = 1973,
                        cadre = True,
                        id = 'ind1',
                        statmarit = u'Célibataire',
                        ),
                    ],
                menages = [
                    dict(
                        personne_de_reference = 'ind0',
                        conjoint = 'ind1',
                    ),
                ],
            ),
            year = 2013,
            ),
        ],
    )
```

```python
request = urllib2.Request('http://api.openfisca.fr/api/1/simulate', headers = {
    'Content-Type': 'application/json',
    'User-Agent': 'OpenFisca-Notebook',
    })
try:
    response = urllib2.urlopen(request, json.dumps(simulation))
except urllib2.HTTPError as response:
    print response.read()
    raise
response_text = response.read()
response_dict = json.loads(response_text)
variables = response_dict['value']
```

```python
scenario = simulation['scenarios'][0]
axes = scenario['axes']
steps_count = 1
for axis in axes:
    steps_count *= axis['count']
persons_count = len(scenario['test_case']['individus'])
total_count = steps_count * persons_count

axes_linspaces = [
    np.linspace(axis['min'], axis['max'], axis['count'])
    for axis in axes
    ]
axes_meshes = np.meshgrid(*axes_linspaces)

sali_list = []
for axis, mesh in zip(axes, axes_meshes):
    array = np.zeros(total_count, dtype = np.float32)
    sali_list.append(list(mesh.reshape(steps_count)))

for i, sali in enumerate(sali_list):
    variables.append(dict(
        code = 'sali{}'.format(i),
        title = 'Salaire imposable de la personne',
        values = sali,
        ))
variables = dict(
    (variable['code'], variable)
    for variable in variables
    )
```
